```python
# 10. Categorical Data Encoding
# Define the function before applying it
def sex(x):
    if x == "female":
        return 0
    else:
        return 1

```

```python
x= df["Sex"].apply(sex)
```

```python
df.insert(5,"Gender",x)
```
